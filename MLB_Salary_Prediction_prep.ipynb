```python
import pandas as pd
```

```python
player_csv = "Resources/People.csv"
salary_csv = "Resources/Salaries.csv"
batting_csv = "Resources/Batting.csv"
fielding_csv = "Resources/Fielding.csv"

player_df = pd.read_csv(player_csv)
salary_df = pd.read_csv(salary_csv)
batting_df = pd.read_csv(batting_csv)
fielding_df = pd.read_csv(fielding_csv)
```

```python
edited_player_df = player_df[["playerID", "nameFirst", "nameLast"]]
edited_salary_df = salary_df[["playerID", "salary", "yearID"]]
edited_salary_df.nunique()
```

```python
#!pip install cpi
import cpi

adjusted_salary = []

for i, row in edited_salary_df.iterrows():
    row_salary = row["salary"]
    row_year = row["yearID"]
    adjusted = cpi.inflate(row_salary, row_year, to = 2016)
    adjusted = round(adjusted)
    adjusted_salary.append(adjusted)

# adjusted_salary

edited_salary_df["Adjusted Salary"] = adjusted_salary
```
